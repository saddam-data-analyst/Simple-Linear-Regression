```python
#Objective
#The objective is to use linear regression to understand how years of experience impact Salary.
```

```python
# Import the libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the salary dataset
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Linear_Regression_Introduction/master/Salary_Data.csv')
```

```python
# Let's have a look :
data.head()
```

```python
# Plotting the Data
#Let’s plot our data points on a 2-D graph to eyeball our dataset(get a rough overview) and see if we can manually find any relationship between the data.
```

```python
plt.plot(data.YearsExperience, data.Salary, 'rx')
plt.show()
```

```python
# There is a clear relationship between the years of experience and salary.
```

```python
# Setting Variable:
X = data[['YearsExperience']]
y = data['Salary']
```

```python
# Spliting the data:
# Next, we split 80% of the data to the training set while 20% of the data to test set using below code. 
# The test_size variable is where we actually specify the proportion of the test set.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Understanding the working of Linear Regression
#The term “linearity” in algebra refers to a linear relationship between two or more variables. 
#If we draw this relationship in a two-dimensional space (between two variables), we get a straight line.

#If we plot the independent variable (x) on the x-axis and dependent variable (y) on the y-axis, 
#linear regression gives us a straight line that "best fits" the data points.
#It’s impossible to connect all the marks with a straight line, so you use a best fitting line.
```

```python
# Training Or Model:
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()  # Create object for the class

# Fit the model to our trainig data i.e learn coefficients
linear_regressor.fit(X_train, y_train)
```

```python
# Interpreting Coefficients
#The coefficients(b0 and b1) will allow us to model our equation with values and find the best fit line. 
#The linear_regressor variable (assigned to a LinearRegression object), is able to have the intercept and coefficients extracted, using the code below.
```

```python
# Print y-intercept
print(linear_regressor.intercept_)

# Print Coefficients
print(linear_regressor.coef_)
```

```python
# Making Predictions Based On your Model:
#Making predictions based on your model is as simple as using the code below: passing the predict method your test data.
#This will return predicted values of y given the new test X data.
```

```python
y_pred = linear_regressor.predict(X_test)  # Make Predictions
```

```python
# We have our predictions in y_pred. Now lets visualize the data set and the regression line with the test data:
```

```python
plt.plot(X_test, y_test, 'rx')
plt.plot(X_test, y_pred, color='black')
plt.show()
```

```python
# Model Evaluations:
#There are three primary metrics used to evaluate linear models.
#These are: Mean absolute error (MAE), Mean squared error (MSE), or Root mean squared error (RMSE).
```

```python
from sklearn import metrics
# Print results of MAE
print(metrics.mean_absolute_error(y_test, y_pred))

# Print results of MSE
print(metrics.mean_squared_error(y_test, y_pred))

# Print result of RMSE
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
